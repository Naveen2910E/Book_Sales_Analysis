```python
import pandas as pd 
import numpy as np 

import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline 

import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('C://Datascience Projects/Kaggle/Resources/Books_Data_Clean.csv')
```

```python
df.head()
```

### Numerical Summary of the Data

```python
df.describe()
```

```python
# Fixing the Negative of Publishing Year 
df = df[df['Publishing Year']>1900]
```

```python
# Generating  descriptive statistics for the DataFrame
df.describe()
```

```python
# Check the number of missing values in each column of the dataframe
df.isna().sum()
```

```python
df.dropna(subset = "Book Name",inplace = True)
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
# Counting the number of unique values in each column of the dataframe
df.nunique()
```

```python
# Create a histogram to visualize the distribution of publishing years 
plt.figure(figsize=(5,4))
plt.hist(df["Publishing Year"])
plt.xlabel("Publishing Year")
plt.ylabel("Frequency")
plt.title("Distribution of Publishing Year")
plt.show()
```

```python
# Create a bar chart visualization showing the distribution of books across different genres
df['genre'].value_counts().plot(kind="bar")
plt.xlabel('genre')
plt.ylabel("Number of Books")
plt.title("Numbers Of Books in Each Genre")
plt.show()
```

```python
# Average rating across all books for each author in the dataset
df.groupby("Author")["Book_average_rating"].mean()
```

```python
# Visualize the distribution of book ratings across different genres using a boxplot
sns.boxplot(x = 'genre', y='Book_average_rating', data=df)
plt.xlabel('genre')
plt.ylabel("Book Ratings Count")
plt.title("Book Ratings count for each Genre")
plt.show()
```

```python
# Scatter plot showing the relationship between sale price and units sold
plt.figure(figsize=(5,4))
plt.scatter(df['sale price'], df['units sold'])
plt.xlabel('Sale Price')
plt.ylabel('Units Sold')
plt.title("Sale Price VS Units Sold")
plt.show()
```

```python
language_counts = df["language_code"].value_counts()
```

```python
# Create a pie chart showing the distribution of books by language
plt.figure(figsize=(5,4))
plt.pie(language_counts, labels=language_counts.index, startangle=90,autopct="%1.1f%%")
plt.title("Language Distribution of Books")
plt.show()
```

```python
# Group by publisher, sum revenues, and sort in descending order to show highest-earning publishers first
df.groupby("Publisher ") ["publisher revenue"].sum().sort_values(ascending=False)
```

```python
df.groupby("Author_Rating") ["Book_ratings_count"].mean().sort_values(ascending = False)
```

```python
df.groupby("language_code").size().sort_values(ascending=False)
```

```python
# scatter plot to visualize the relationship between book average ratings and number of ratings
plt.figure(figsize=(6,5))
plt.scatter(df["Book_average_rating"],df["Book_ratings_count"])
plt.xlabel("Book_average_rating")
plt.ylabel("Book_ratings_count")
plt.title(" Scatterplot of Book average rating vs Book ratings count")
plt.show()
```

```python
# Calculate the total gross sales for each author by grouping the dataframe by "Author" column
total_gross_sales_by_author = df.groupby("Author")["gross sales"].sum()
```

```python
# Create a bar plot showing the top 10 authors by total gross sales
total_gross_sales_by_author.sort_values(ascending = False).head(10).plot(kind="bar")
plt.xlabel("Author")
plt.ylabel("Total gross sales")
plt.title("Tota; gross sales of each author")
plt.show()
```

```python
df.groupby("Publishing Year") ["units sold"].sum().plot(kind="line")
plt.xlabel("Publishing Year")
plt.ylabel("Total units sold")
plt.title("Total units sold over the years")
plt.show()
```
